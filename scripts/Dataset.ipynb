```python
import pandas as pd
import os
from os import walk

import skimage
from skimage import io
```

```python
mypath = "/media/Store/juan/Datasets/VOC 2012/VOCdevkit/VOC2012/JPEGImages"

res = []
for (root, dirnames, files) in walk(mypath):
    if len(files) > 0:
        res.extend(list(zip([root]*len(files), files)))

df = pd.DataFrame(res, columns=['Path', 'File_Name'])

df = df[(df['File_Name'].str.contains("jpg"))]

print(df.describe())
df.head()
```

```python
dataset = df.copy()

# Read image size

dataset['Size'] = dataset.apply(lambda row: os.path.join(row.Path, row.File_Name), axis=1)

dataset['Size'] = dataset.apply(lambda row: skimage.io.imread(row['Size']).shape, axis=1)

dataset[['high','width', 'channels']] = dataset['Size'].apply(pd.Series)

dataset.head()
```
