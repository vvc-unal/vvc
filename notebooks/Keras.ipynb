# Enviroment setup

## Requirements

* Ubuntu 16.04
* Nvidia Graphic card, GTX 750 ti or above

### Software

* Nvidia drivers, 384.xx
* Nvidia CUDA 8
* NVIDIA cuDNN 6
* Python 2.7
* Tensorflow 1.3.0 [https://www.tensorflow.org/install/install_linux]
* Keras 2.0.8

## Instructions

1. Install Ubuntu 16.04
2. Nvidia drivers (with binary ``nvidia-modprobe``)
3. Nvidia CUDA 8
    1. 
    ```bash
   
        sudo dpkg -i cuda-repo-ubuntu1604_8.0.61-1_amd64.deb
        sudo apt-get update
        sudo apt-get install cuda

    ```
    2. 
     ```bash
     export LD_LIBRARY_PATH=/usr/local/cuda-8.0/lib64\
                         ${LD_LIBRARY_PATH:+:${LD_LIBRARY_PATH}}
     ```
4. NVIDIA cuDNN
    1. Download and uncompress it.
    2. Set variable
    ```bash
   export LD_LIBRARY_PATH="/home/sebas/Programas/cuda/lib64:/usr/local/cuda-8.0/lib64"
   ```
5. Install Anaconda
6. Create virtual enviroment
    ```bash
       conda create -n py27-gpu python=2.7
        
       source activate py27-gpu
       
       conda install pillow h5py
       
       # If you want to use jupyter notebooks with different enviroments
       conda install nb_conda
              
       conda install -c menpo opencv3
       
       conda remove numpy
       
       pip install numpy
       
       pip install opencv-python

    ```
7. TensorFlow and Keras
    ```bash
    
        # library for video reading and writing
        pip install sk-video
    
        pip install --ignore-installed --upgrade tensorflow-gpu
    
        pip install keras
    ```

## Instructions (Docker)

Instructions to install Tensorflow and Keras with Docker. See [TensorFlow with Docker](https://www.tensorflow.org/install/install_linux#InstallingDocker).

1. Install Ubuntu 16.04
1. Install Nvidia drivers 384
1. Install Docker
1. Install Nvidia-docker (2.0)
4. Create Tensorflow container:
    ```bash
   nvidia-docker run -it tensorflow/tensorflow:latest-gpu bash
   ```
5. Install Anaconda
6. Create virtual enviroment
    ```bash
       conda create -n py27-gpu python=2.7
        
       source activate py27-gpu
       
       conda install pillow h5py
       
       # If you want to use jupyter notebooks with different enviroments
       conda install nb_conda
              
       conda install -c menpo opencv3
       
       conda remove numpy
       
       pip install numpy
       
       pip install opencv-python

    ```
7. [TensorFlow](https://www.tensorflow.org/install/install_linux#InstallingDocker)
    ```bash
    
        # library for video reading and writing
        pip install sk-video
    
        pip install --ignore-installed --upgrade tensorflow-gpu
    
        pip install keras
    ```
